```python
import pandas as pd
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool

df  = pd.read_csv('drawpython.csv')
source = ColumnDataSource(df)
p = figure()
p.circle(x='x', y='y',
         source=source,
         size=10, color='green')
p.title.text = 'A Sample Scatter Chart'
p.xaxis.axis_label = 'X Values'
p.yaxis.axis_label = 'Y Values'
hover = HoverTool()
hover.tooltips=[
    ('X Value', '@x'),
    ('Y Value', '@y')
]
p.add_tools(hover)
show(p)
```
